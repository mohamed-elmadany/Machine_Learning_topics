```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np
import pandas as pd
```

```python
data=load_iris()
```

```python
x=data.data
y=data.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=00.3,random_state=42)
```

```python
params={
    'C':[.1,1,10],
    'kernel':['linear','rbf'],
    'gamma':['scale','auto']
}
```

```python
model=SVC()
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
grid=GridSearchCV(
    estimator=model,
    param_grid=params,
    scoring='accuracy',
    cv=5,
    n_jobs=-1
)
```

```python
grid.fit(x_train,y_train)
```

```python
res=pd.DataFrame(grid.cv_results_)
```

```python
res
```

```python
res[['params','mean_test_score','std_test_score']]
```

```python
grid.best_params_
```

```python
randomS=RandomizedSearchCV(
    estimator=model,
    param_distributions=params,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    n_iter=10
)
```

```python
randomS.fit(x_train,y_train)
```

```python
res2=pd.DataFrame(randomS.cv_results_)
```
